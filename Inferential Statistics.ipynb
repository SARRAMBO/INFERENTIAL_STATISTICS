```python
#CHI SQUARE TEST- This test is applied when you have two categorical variables from a single population.Non parametric test performed on categorical data.
```

```python
import scipy.stats as stats #all test(ztf present in stats library)
import seaborn as sns
import pandas as pd
import numpy as np
dataset=sns.load_dataset('tips')
print(dataset)
```

```python
#AIM - to evaluate relationship between two categorical variables,ie,smoking and gender.
#Ho(null hypothesis)- smoking and gender are not related
#H1(alternate hypothesis)- smoking and gender related.

#PRINTING OBSERVED VALUES-
dataset_table=pd.crosstab(dataset['sex'],dataset['smoker']) #prints matrix concerning relationship between two categorical variables
print(dataset_table)
Observed_Values = dataset_table.values  #printing out observed values depicting relationship based on the data.
```

```python
#Deriving expected values-(from observed values)
val=stats.chi2_contingency(dataset_table)
print(val)

#formula for deriving expected value matrix - fe = (fo *frow)/ n
#fe- expected value at a given cell (for ex fe(male,yes))
#fo- observed value at a given cell (for ex- fo(male,yes)=60)
#frow - value obtained by adding all entries present in the row= 60+97 = 157
#n - total no of observations= 157+84 = 241
#fe(Male,yes) = (60 * 157)/241

Expected_values = val[3] #calculated expected values
print(Expected_values)
```

```python
#calculating degrees of freedom from rows and columns-
no_of_rows=len(dataset_table.iloc[0:2,0])
no_of_columns=len(dataset_table.iloc[0,0:2])

ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05 #state hypothesis with 95% confidence interval.
```

```python
#chisquare statistic calculation-
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_values)]) #(fo - fe ) 2 / fe
print(chi_square)
chi_square_statistic=chi_square[0]+chi_square[1] #to be used later in computing p value.
```

```python
#deriving critcial chisquare value from chisquare table - q = confidence interval df=degree of freedom
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
#calculating threshold chisquare value(critical value) from chisquare table based on  confidence interval/significance value and degree of freedom
```

```python
# using chisquare statistic to compute p-value based on dof,chisquare statistic-
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
#chi2.cdf- used to calculate the confidence interval/pvalue.
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('critical_value:',critical_value)
print('chisquare_statistic:',chi_square_statistic)
```

```python

#STATING DECISION RULE-
#On basis of chisquare statistic-(plotting rejection region)
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")

#On basis of p value-(preferred for coding)
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

```python
#ONE SAMPLE T-TEST WITH PYTHON-
ages=[10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,16,17,32,35,26,27,65,18,43,23,21,20,19,70]
len(ages)
#POPULATION MEAN-
import numpy as np
ages_mean=np.mean(ages)
print(ages_mean)
```

```python
#SAMPLING POPULATION WITH A SAMPLE SIZE=10
sample_size=10
age_sample=np.random.choice(ages,sample_size)
age_sample
```

```python

from scipy.stats import ttest_1samp

ttest,p_value=ttest_1samp(age_sample,30)
print(p_value)
```

```python
#STATING DECISION RULE-
#accepting or rejecting based on p value-
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

```python
#TWO SAMPLE T test-
#compares samples of two independent groups to arrive at conclusion based on their population parameters (mean)

import numpy as np
import pandas as pd
import scipy.stats as stats
import math
np.random.seed(6)
classA_ages=stats.poisson.rvs(loc=18,mu=30,size=60)
np.random.seed(12)
ClassB_ages=stats.poisson.rvs(loc=18,mu=33,size=60)
ClassB_ages.mean()
```

```python
#STATING HYPOTHESIS-
#Ho- difference between mean of two groups is not statistically significant.
#H1- difference between mean of two groups is statistically significant.
```

```python
#STATING DESCISION RULE-
_,p_value=stats.ttest_ind(a=classA_ages,b=ClassB_ages,equal_var=False)
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
p_value #falling below significance level
```

```python
#PAIRED T-TEST-
#When you want to check how different samples from the same group are, you can go for a paired T-test

weight1=[25,30,28,35,28,34,26,29,30,26,28,32,31,30,45]
weight2=weight1+stats.norm.rvs(scale=5,loc=-1.25,size=15)
print(weight1)
print(weight2)
weight_df=pd.DataFrame({"weight_10":np.array(weight1),
                         "weight_20":np.array(weight2),
                       "weight_change":np.array(weight2)-np.array(weight1)})
weight_df
#STATING HYPOTHESIS-
#Ho- difference between mean of two samples is not statistically significant(samples are similar)
#H1- difference between mean of two samples is statistically significant(samples are not similar)

#STATING DECISION RULE-
_,p_value=stats.ttest_rel(a=weight1,b=weight2)
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

```python
#ANNOVA TEST-(One way F Test)
#The t-test works well when dealing with two groups, but sometimes we want to compare more than two groups at the same time.
#For example, if in a iris dataset,we wanted to test whether petal_width age differs based on some categorical variable like species, we have to compare the means of each level or group the variable
#Example : there are 3 different category of iris flowers and their petal width and need to check whether all 3 group are similar or not
import seaborn as sns
df1=sns.load_dataset('iris')
df1.head()
```

```python
#determining no of distinct species-
df_anova = df1[['petal_width','species']]
grps = pd.unique(df_anova.species.values)
grps
 #separating entries of diff species in a dictionary-
d_data = {grp:df_anova['petal_width'][df_anova.species == grp] for grp in grps}
print(d_data)
```

```python
#STAtING HYPOTHESIS-
#Ho(null hypothesis)- Means are similar between 3 groups(difference not statistically significant)
#H1(alternative hypothesis)- Means are not similar between 3 groups(diff is statistically significant)
```

```python
#CALCULATING P VALUE-
F, p = stats.f_oneway(d_data['setosa'], d_data['versicolor'], d_data['virginica'])
print(p)
```

```python
#STATING DECISION RULE-
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
